```python
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

fig = plt.figure(figsize=(7,5))
ax = plt.axes()

ax.plot(x,np.sin(x),'-',label='f(x) = sin()')
ax.plot(x,np.cos(x),'--',label='f(x) = cos()')

# ax.set()

plt.axis([0,10,1,-1])
# plt.axis('tight')

plt.title('Smaple')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# plt.close()
```
